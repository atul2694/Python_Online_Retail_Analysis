## Task 1: Load the Data

```python
# your code here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Modules are imported")
```

```python
df=pd.read_excel("Online Retail.xlsx")
```

```python
df.head()
```

```python
## Task 2 :Data Cleaning
```

```python
#Handle Missing Values 
#Check for missing values 
df.isnull().sum()
```

```python
# Remove Missing rows 
df=df.dropna(subset=["CustomerID","Description"])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Remove Duplicate 
# Check for Duplicates:

df.duplicated().sum()
```

```python
#Remove Duplicates
df= df.drop_duplicates()
```

```python
df.shape
```

```python
#Remove negative 
# Check for Negative 
df[df["Quantity"]<=0]
```

```python
# Remove negative 
df=df[df["Quantity"]>0]
```

```python
df.shape
```

```python
# convrt invoicedate column to datetime format 
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
df.dtypes
```

```python
# Basic Statatics
df.describe()
```

```python
# measures of Central Tendancy 
# Mean
mean_quantity = df['Quantity'].mean()
mean_price = df['UnitPrice'].mean()

# Median
median_quantity = df['Quantity'].median()
median_price = df['UnitPrice'].median()

# Mode
mode_quantity = df['Quantity'].mode()[0]
mode_price = df['UnitPrice'].mode()[0]
```

```python
# Measure of Dispeprsion 
# Standard deviation
std_quantity = df['Quantity'].std()
std_price = df['UnitPrice'].std()

# Variance
var_quantity = df['Quantity'].var()
var_price = df['UnitPrice'].var()

# Range
range_quantity = df['Quantity'].max() - df['Quantity'].min()
range_price = df['UnitPrice'].max() - df['UnitPrice'].min()
```

```python
#Skewness and Kurtosis 
df['Quantity'].skew()
df['Quantity'].kurt()

df['UnitPrice'].skew()
df['UnitPrice'].kurt()
```

```python
# Add Total Price Column 
df["TotalPrice"]=df["Quantity"]*df["UnitPrice"]
```

```python
df.head()
```

```python
#Data Visualization 

# 1. Histogram of Quantity  
plt.figure(figsize=(10,5))
plt.hist(df["Quantity"],bins=50,color="skyblue",edgecolor="black")
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')

plt.show()

# Histogram of Unit Price
plt.figure(figsize=(10,5))
plt.hist(df["UnitPrice"],bins=50,color="skyblue",edgecolor="black")
plt.title('Distribution of Unit Price')
plt.xlabel('UnitPrice')
plt.ylabel('Frequency')

plt.show()
```

```python

#scatter plot :Quantity Vs UnitPrice
plt.figure(figsize=(10,5))
sns.scatterplot(x="Quantity",y="UnitPrice",data=df)
plt.title("Quantity Vs UnitPrice")
plt.xlabel("Quantity")
plt.ylabel("UnitPrice")
plt.xlim(0,100)
plt.ylim(0,100)
plt.show()
```

```python
# Top 10 products by revunue 

top_10_products=df.groupby("Description")["TotalPrice"].sum().sort_values(ascending=False).nlargest(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_10_products.values, y=top_10_products.index,hue=top_10_products.index,palette="viridis")
plt.title("Top 10 Product By Revunue")
plt.xlabel("Total revunue")
plt.ylabel("Top 10 Product")
plt.tight_layout()

plt.show()
```

```python
# monthly Revunue Trend

df["Month"]=df["InvoiceDate"].dt.to_period("M")
monthly_revunue=df.groupby("Month")["TotalPrice"].sum().reset_index()
monthly_revunue["Month"]=monthly_revunue["Month"].astype(str)

plt.figure(figsize=(10,6))
sns.lineplot(data=monthly_revunue,x="Month",y="TotalPrice",marker="o")
plt.title("Monthly Revnue Trends")
plt.xticks(rotation=90)
plt.ylabel("Revunue")
plt.tight_layout()
plt.savefig("Monhtly_revune_trend.png")
plt.show()
```

```python
# Revunue by Country 
revunue_by_country=df.groupby("Country")["TotalPrice"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=revunue_by_country.values,y=revunue_by_country.index,hue=revunue_by_country.index,palette="magma")
plt.title("Revunue by Country")
plt.xlabel("Total Revunue")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("Revunue_By_Country")
plt.show()
```

```python
# Sales trends over Time 

#Extract Time Feature
df['Month'] = df['InvoiceDate'].dt.month
df['Year'] = df['InvoiceDate'].dt.year
df['Weekday'] = df['InvoiceDate'].dt.day_name()
```

```python
# Monthly sales Trends
df["Month"]=df["InvoiceDate"].dt.to_period("M")
monthly_revunue=df.groupby("Month")["TotalPrice"].sum().reset_index()
monthly_revunue["Month"]=monthly_revunue["Month"].astype(str)

plt.figure(figsize=(10,6))
sns.lineplot(data=monthly_revunue,x="Month",y="TotalPrice",marker="o")
plt.title("Monthly Revnue Trends")
plt.xticks(rotation=90)
plt.ylabel("Revunue")
plt.tight_layout()
plt.savefig("Monhtly_revune_trend.png")
plt.show()
```

```python
#Busiest Day of the week
weekday_sales = df.groupby('Weekday')['TotalPrice'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=weekday_sales.index, y=weekday_sales.values, palette='coolwarm')
plt.title('Revenue by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top-Selling Products by Quantity
top_selling_producy_by_quantity=df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_selling_producy_by_quantity.values,y=top_selling_producy_by_quantity.index,hue=top_selling_producy_by_quantity.index,palette="magma")
plt.title("Top Selling Product By Quantity")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Producty Description")
#plt.legend(title="Product",bbox_to_anchor=(1,1))
plt.tight_layout()
plt.show()
```

```python
# Top Countries by Quantity
top_Country_by_quantity=df.groupby("Country")["Quantity"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_Country_by_quantity.values,y=top_Country_by_quantity.index,hue=top_Country_by_quantity.index,palette="magma")
plt.title("Top Countres By Quantity")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
